```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x, y)
plt.xlabel('x axis')
plt.ylabel('y lable')
plt.title(" Basic Line Plot")


```

```python


plt.plot(x, y, color='red', linestyle='-.', linewidth=3, marker='o', markersize=9)

# Add grid
plt.grid(True)

# Optional: Add title and axis labels
plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```

```python

x = [1, 2, 3, 4, 5]
y1 = [i**2 for i in x]        # y1 = x^2  [1, 4, 9, 16, 25]
y2 = [30 - 5*i for i in x]    # y2 = 30 - 5x  [25, 20, 15, 10, 5]

plt.figure(figsize=(9, 5))  # Correct usage

plt.subplot(2, 2, 1)  # First subplot

# Plot two lines that will cross
plt.plot(x, y1, color='green', label='y = x')
plt.plot(x, y2, color='red', label='y = 30 - 5x')

plt.title('Crossing Lines')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()




```
